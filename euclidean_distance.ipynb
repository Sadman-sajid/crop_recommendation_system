```python
s_crop=[[0,19,8,15,8,0.5,0.5],[1,16,10,16,3,0.3,0],[2,19,3,13,5,0,0],[3,3,3,3,2,0,0],[4,15,5,10,3,0,0],[5,3,5,6,2,0,0],[6,15,6,13,3,0,0],[7,3,5,6,3,0,0]]
print(s_crop)
```

```python
crops={
    0:'Rosun',
    1:'Morich',
    2:'Paat',
    3:'Kheshari',
    4:'Shorisha',
    5:'Borboti',
    6:'Tula',
    7:'Mugg',
}

for ind in soil_data.index:

  best=99;
  sec_best=99;
  third_best=99;

  ns=soil_data['N'][ind]
  ps=soil_data['P'][ind]
  ks=soil_data['K'][ind]
  ss=soil_data['S'][ind]
  zs=soil_data['Z'][ind]
  bs=soil_data['B'][ind]

  arr=[];
  sortedarr=[];

  for j in range (0,8):

    nc=s_crop[j][1]
    pc=s_crop[j][2]
    kc=s_crop[j][3]
    sc=s_crop[j][4]
    zc=s_crop[j][5]
    bc=s_crop[j][6]

    euc=((ns-nc)**2)+((ps-pc)**2)+((ks-kc)**2)+((ss-sc)**2)+((zs-zc)**2)+((bs-bc)**2)
    euc=math.sqrt(euc)

    arr.append(euc)
    
  for k in range(0,8):
    sortedarr.append(arr[k])

  sortedarr.sort()  


  for a in range(0,3):
    for b in range(0,8):
      if(sortedarr[a]==arr[b]):
        if(a==0):
          best=b;
        elif(a==1):
          sec_best=b;
        elif(a==2):
          third_best=b;

  print(ind+2, soil_data['Mrittika Dol'][ind] ,soil_data['Bhumi'][ind], crops[best])
  #print(ind+2, soil_data['Mrittika Dol'][ind] ,soil_data['Bhumi'][ind], crops[sec_best])
  #print(ind+2, soil_data['Mrittika Dol'][ind] ,soil_data['Bhumi'][ind], crops[third_best])

  #ws.append([soil_data['Mrittika Dol'][ind],soil_data['Bhumi'][ind], soil_data['Bunot'][ind],soil_data['N'][ind], soil_data['P'][ind],soil_data['K'][ind], soil_data['S'][ind], soil_data['Z'][ind],soil_data['B'][ind],crops[best]])
  #ws.append([soil_data['Mrittika Dol'][ind],soil_data['Bhumi'][ind], soil_data['Bunot'][ind],soil_data['N'][ind], soil_data['P'][ind],soil_data['K'][ind], soil_data['S'][ind], soil_data['Z'][ind], soil_data['B'][ind], crops[sec_best]])
  #ws.append([soil_data['Mrittika Dol'][ind],soil_data['Bhumi'][ind], soil_data['Bunot'][ind],soil_data['N'][ind], soil_data['P'][ind],soil_data['K'][ind], soil_data['S'][ind], soil_data['Z'][ind], soil_data['B'][ind], crops[third_best]])

  
 
 
#wk.save('/content/drive/MyDrive/main(xslx).xlsx')
```
